<p style="font-size: 17px">9. Исходный код программы.</p>

```python
ZERO = ord('0')
NINE = ord('9')
PLUS = ord('+')
MINUS = ord('-')
X = ord('x')
POWER = ord('^')
STR_END = '$'


class Parser:
    def __init__(self, expr: str):
        self.expr = expr

    def parse(self):
        self._ABS_POLYNOMIAL(self.expr + STR_END)

    def _ABS_POLYNOMIAL(self, expr: str):
        sign = self._SIGN(expr)
        if sign:
            self._POLYNOMIAL(sign)
            return
        condition = self._CONDITION(expr)
        if condition:
            self._POLYNOMIAL(condition)

    def _POLYNOMIAL(self, expr: str):
        if expr[0] == STR_END:
            return True
        sign = self._SIGN(expr)
        if sign:
            self._POLYNOMIAL(sign)
        else:
            raise ValueError

    def _SIGN(self, expr: str):
        if not (ord(expr[0]) == PLUS or ord(expr[0]) == MINUS):
            return False
        condition = self._CONDITION(expr[1:])
        if condition:
            return condition
        return False

    def _CONDITION(self, expr: str):
        base = self._BASE(expr)
        if base:
            return base
        number = self._NUMBER(expr)
        number_base = self._BASE(number)
        if number_base:
            return number_base
        if number:
            return number
        raise ValueError

    def _BASE(self, expr: str):
        if not expr or ord(expr[0]) != X:
            return False
        power = self._POWER(expr[1:])
        if power:
            return power
        return expr[1:]

    def _POWER(self, expr: str):
        if not expr or ord(expr[0]) != POWER:
            return False
        number = self._NUMBER(expr[1:])
        if number:
            return number
        return False

    def _NUMBER(self, expr: str):
        i = 0
        while ZERO <= ord(expr[i]) <= NINE:
            i += 1
        return expr[i:] if i != 0 else False


if __name__ == '__main__':
    expr = input('Enter expression: ')
    parser = Parser(expr)
    try:
        parser.parse()
        print('Ok - expression is right')
    except ValueError as e:
        print('Error in expression - wrong expression')
```
